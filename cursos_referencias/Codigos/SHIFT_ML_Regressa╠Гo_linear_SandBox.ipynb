```python
import matplotlib.pyplot as plt
import numpy as np

import statsmodels.api as sm

from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
```

```python
x,y = make_regression(n_samples = 130, bias = 1.0, noise = 20,n_features=1)

plt.scatter(x,y)
```

```python
linreg = LinearRegression().fit(x,y)

print('linear model coeff (w): {}'.format(linreg.coef_))
print('linear model intercept (b): {:.3f}'.format(linreg.intercept_))
print('R-squared score (training): {:.3f}'.format(linreg.score(x,y)))


plt.scatter(x,y, marker= 'o', s=50)
plt.plot(x, linreg.coef_ * x + linreg.intercept_, 'r-')
plt.show()
```

CRIANDO OUTRO CONJUNTO DE DADOS


```python
x_min = 0
x_max = 10
amostras = (x_max-x_min)*10
ruido = 1

x = np.random.rand(amostras) * (x_max-x_min) + x_min
x= np.sort(x)

y = np.cos(x)
y += (np.random.rand(amostras)-.5) * ruido

plt.scatter(x,y)

```

```python
p = np.polynomial.Chebyshev.fit(x,y,5)

plt.plot(x, y, 'b.')
plt.plot(x, p(x), 'r-')
plt.show()
```
