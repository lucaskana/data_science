```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import cm
from collections import OrderedDict
from sklearn.cluster import KMeans

%load_ext google.colab.data_table
```

```python
X = np.load('../content/DBSCAN - base01.npy')
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
min_clusters = 1
max_clusters = 10

wcss = []
for i in range(min_clusters, max_clusters):
  kmeans = KMeans(n_clusters=i)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

plt.plot(range(min_clusters, max_clusters),wcss)
plt.show()
```

```python
clusters =  8

kmeans = KMeans(n_clusters=clusters)
y = kmeans.fit_predict(X)

fig = plt.scatter(X[:,0],X[:,1])
fig = plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=150,marker= 'X', c='red')
plt.show()
```

```python
labels = kmeans.labels_

fig = plt.scatter(X[:,0],X[:,1],c=labels)
fig = plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=150,marker= 'X', c='red')
plt.show()
```

```python
from scipy.spatial import Voronoi, voronoi_plot_2d

vor = Voronoi(kmeans.cluster_centers_)
fig = voronoi_plot_2d(vor,show_vertices=False,line_colors='red',show_points=False)
fig = plt.scatter(X[:,0],X[:,1],c=labels)
fig = plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=150,marker= 'X', c='red')
plt.axis([2,6,2,6])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target

plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
min_clusters = 1
max_clusters = 10

wcss = []
for i in range(min_clusters, max_clusters):
  kmeans = KMeans(n_clusters=i)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

plt.plot(range(min_clusters, max_clusters),wcss)
plt.show()
```

```python
clusters =  3

kmeans = KMeans(n_clusters=clusters)
y = kmeans.fit_predict(X)
labels = kmeans.labels_
fig = plt.scatter(X[:,0],X[:,1], c=labels)
fig = plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=150,marker= 'X', c='red')
plt.show()
```

```python
from scipy.spatial import Voronoi, voronoi_plot_2d

vor = Voronoi(kmeans.cluster_centers_)
fig = voronoi_plot_2d(vor,show_vertices=False,line_colors='red',show_points=False)
fig = plt.scatter(X[:,0],X[:,1],c=labels)
fig = plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=150,marker= 'X', c='red')
plt.axis([2,6,2,6])
plt.show()
```
