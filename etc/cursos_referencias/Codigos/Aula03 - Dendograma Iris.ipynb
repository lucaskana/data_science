```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np


iris = datasets.load_iris()
```

```python
feat = iris.feature_names
X = iris.data
```

```python
figure = plt.figure(figsize=(20,8))
Z = linkage(X,method='average',optimal_ordering=True)
dendrogram(Z,p=3,truncate_mode='level',distance_sort=True)
plt.title('DENDOGRAMA - Cluster Hierárquico')
plt.xlabel('Index do ponto ou Quantidade de pontos se estiver entre parenteses\n ')
plt.ylabel('Distância Euclidiana')
plt.tight_layout()
plt.show()
```

```python
Zz = np.array_str(Z, precision=4, suppress_small=True)
print(len(Zz))
print(Zz)
```
