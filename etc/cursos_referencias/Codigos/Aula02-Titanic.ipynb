```python
import pandas as pd
from pandas import read_csv
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

dataset = read_csv('https://telescopeinstorage.blob.core.windows.net/datasets/titanic-apriori.csv', sep=';' , engine='python')
dataset.head()
```

```python
dataset_encoded = pd.get_dummies(dataset[['Class','Sex','Age','Survived']])
dataset_encoded
```

```python
frequencias = apriori(dataset_encoded,use_colnames=True, min_support=0.1)
print(frequencias)
```

```python
regras = association_rules(frequencias, metric='confidence', min_threshold=0.4)

regras = regras.sort_values ('lift',ascending=False)
regras = regras[['antecedents','consequents','support','confidence', 'lift' ]]

regras
```
