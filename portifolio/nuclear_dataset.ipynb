# Importando MÃ³dulos / Download Dados

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import matplotlib.dates as mdates
import matplotlib as mpl
import kaggle
import os
```

```python
def downloadKaggle(fName, pathFile,kaggleFile):
    kaggle.api.authenticate()
    kaggle.api.dataset_download_files(fName, path=pathFile, unzip=True)
    dirname = os.path.abspath('')
    arquivo=dirname + '/../datasets/'+kaggleFile
    return arquivo


dataset = 'liananapalkova/nuclear-power-plants'
data_file_name = 'energy-pop-exposure-nuclear-plants-locations_plants.csv'


arquivo = downloadKaggle(dataset,'../datasets',data_file_name)
```

```python
df=pd.read_csv(arquivo,sep=',')

df.head(150)
```
